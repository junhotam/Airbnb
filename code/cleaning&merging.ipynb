# Merging & Cleaning Dataset

### 00 | import libraries and data

```python
import pandas as pd
import numpy as np
import os
```

```python
path = r'/Users/johntam/cf_data_analysis/tasks/A6/code/raw_data'

cities = ['amsterdam', 'athens', 'barcelona', 'berlin', 'budapest', 'lisbon', 'london', 'paris', 'rome', 'vienna']

data = {}

for city in cities:
    data[str(city) + '_weekends'] = pd.read_csv(os.path.join(path, city + '_weekends.csv'))
    data[str(city) + '_weekdays'] = pd.read_csv(os.path.join(path, city + '_weekdays.csv'))
```

### 01 | merging data

```python
# combining the weekend and weekday datasets for each city

def combine(weekdays,weekends,city):       
    data[weekdays]['weekends'] = 0
    data[weekends]['weekends'] = 1
    data[weekdays].drop(columns = ['Unnamed: 0'],inplace=True)
    data[weekends].drop(columns = ['Unnamed: 0'],inplace=True)
    merged = pd.concat([data[weekdays], data[weekends]])
    merged['city'] = city
    return merged
```

```python
amsterdam = combine('amsterdam_weekdays','amsterdam_weekends','amsterdam')
athens = combine('athens_weekdays','athens_weekends','athens')
barcelona = combine('barcelona_weekdays','barcelona_weekends','barcelona')
berlin = combine('berlin_weekdays','berlin_weekends','berlin')
budapest = combine('budapest_weekdays','budapest_weekends','budapest')
lisbon = combine('lisbon_weekdays','lisbon_weekends','lisbon')
london = combine('london_weekdays','london_weekends','london')
paris = combine('paris_weekdays','paris_weekends','paris')
rome = combine('rome_weekdays','rome_weekends','rome')
vienna = combine('vienna_weekdays','vienna_weekends','vienna')
```

```python
cities = [amsterdam, athens, barcelona, berlin, budapest, lisbon, london, paris, rome, vienna]

eu_data = pd.concat(cities, ignore_index=True) 
```
